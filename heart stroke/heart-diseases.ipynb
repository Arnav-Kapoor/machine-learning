```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
dataset=pd.read_csv('/kaggle/input/stroke-prediction-dataset/healthcare-dataset-stroke-data.csv')
```

```python
df=pd.DataFrame(dataset)
```

```python
df
```

```python
df.drop(['id','ever_married'],axis=1,inplace=True)
```

```python
df.head()
```

```python
df.info()
```

```python
print(np.unique(df['gender']),np.unique(df['work_type']),np.unique(df['Residence_type']),np.unique(df['smoking_status']))
```

```python
df=pd.get_dummies(df,columns=['gender','work_type','Residence_type','smoking_status'],drop_first=True)
```

```python
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df['bmi'].median()
```

```python
df['bmi']=df['bmi'].fillna(df['bmi'].median())
```

```python
df.isnull().sum()
```

```python
df.info()
```

```python
df['heart_disease'].value_counts()
```

```python
df['stroke'].value_counts()
```

```python
X=df.drop('stroke',axis=1)
```

```python
y=df['stroke']
```

```python
print(X.shape,y.shape)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=22)
```

```python
print(X_train.shape,y_train.shape,X_test.shape,y_test.shape)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dtc=DecisionTreeClassifier()
```

```python
dtc.fit(X_train,y_train)
```

```python
preds=dtc.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix,classification_report
```

```python
confusion_matrix(preds,y_test)
```

```python
print("accuracy: ",(1373/(1373+12+76+72))*100)
```

```python
print(classification_report(preds,y_test))
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rf.fit(X_train,y_train)
```

```python
preds_rf=rf.predict(X_test)
```
